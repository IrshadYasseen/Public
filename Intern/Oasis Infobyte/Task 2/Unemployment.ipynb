#### KDE plot for Hypothysis testing

- The kde plot show how much value we find at a give point the higher the y value the higher the probability of finding the value. Used in calculating data spread estimating the probability density of data points. 

- Peaks in the KDE plot represent dates where there are many observations, i.e., these dates occur frequently in your data

- Valleys or low points represent dates where there are fewer observations, i.e., these dates occur less frequently.

- If there are multiple peaks, it indicates that your data may have multiple modes (i.e., multiple dates that occur frequently)

- Also used for finding outliers

```python
plt.figure(figsize=(8, 5))
sns.kdeplot(data=df['Date'],fill=True)
plt.title("Kernel Density Estimation of Dates")
plt.xlabel("Date")
plt.ylabel("Density")
```

#### Strip plot

- A Strip Plot is a bit similar to a Swarm Plot and displays individual data points along a single axis. However, the
data points will overlap in the strip plot, so we can say that Swarm Plots are particularly useful with smaller
datasets and when you want to prevent overlap.

- Use sns.stripplot(data=dataframe[column]) , and you can also use this for bivariate. Jittering (adding a small
amount of random noise) can be applied to the data points to reduce the overlap and improve visibility

#### Removing Outliers

```python
def remove_outliers(column):
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return column[(column >= lower_bound) & (column <= upper_bound)]


df['Estimated Unemployment Rate (%)'] = remove_outliers(df['Estimated Unemployment Rate (%)'])



df['ELP%'] = remove_outliers(df['ELP%'])

df['Estimated Employed'] = remove_outliers(df['Estimated Employed'])

df.reset_index(drop=True, inplace=True)
```
